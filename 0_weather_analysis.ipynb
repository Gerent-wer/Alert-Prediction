```python
data_grouped_by_city = data.groupby(by=["city_address"]).mean()
```

```python
data_grouped_by_city
```

```python
# as we can see, the avarage temperatues also are a bit different from city to city
plt.bar(data_grouped_by_city.index,data_grouped_by_city['day_temp'] )
plt.xticks(rotation=90)
plt.title('Average temperature by Ukraine cities')
plt.xlabel('City')
plt.ylabel('Avg temparature')
plt.show()
```

```python
# as we can see, the average wind speeds are a bit various, so we can probably pay more attention on this feature 
# considering the prediction of a new alarm

plt.bar(data_grouped_by_city.index,data_grouped_by_city['day_windspeed'] )
plt.xticks(rotation=90)
plt.title('Average wind speed by Ukraine cities')
plt.xlabel('City')
plt.ylabel('Avg windspeed')
plt.show()
```

```python
# as we can see, the average humidities are almost the same through the cities

plt.bar(data_grouped_by_city.index,data_grouped_by_city['day_humidity'] )
plt.xticks(rotation=90)
plt.title('Average humidity by Ukraine cities')
plt.xlabel('City')
plt.ylabel('Avg humidity')
plt.show()
```

```python
# Visibility of Kharkiv ciry seems actually quite strange, very low visibility

plt.bar(data_grouped_by_city.index,data_grouped_by_city['day_visibility'] )
plt.xticks(rotation=90)
plt.title('Average visibility by Ukraine cities')
plt.xlabel('City')
plt.ylabel('Avg visibility')
plt.show()
```
