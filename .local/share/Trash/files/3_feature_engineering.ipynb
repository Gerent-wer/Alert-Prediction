```python
data["start_time"] = pd.to_datetime(data["start"])
data["end_time"] = pd.to_datetime(data["end"])
```

```python
# fisrt feature a number of regions where the alarm is going on

for i in range(0,data.shape[0]):
    count = 0
    for j in range(0,data.shape[0]):
        if (data['start_time'][j] >= data['start_time'][i]) and (data['end_time'][j] < data['end_time'][i]):
            count = count + 1
    data["feature_number_of_regions"][i] = count
    count = 0
```
