```python
import pandas as pd
```

```python
data1 = pd.read_csv('df_events_v4.csv')
data1 = data1.drop(columns=['Unnamed: 0'],axis=1)
```

```python
data2 = pd.read_csv('df_weather_reg.csv')
data2 = data2.drop(columns=['Unnamed: 0'],axis=1)
```

```python
data1.head()
```

```python
data2.head()
```

```python
df_weather_v4 = data2.merge(data1, 
                                  how="left", 
                                  left_on=["region_alt","hour_datetimeEpoch"],
                                  right_on=["event_region_title","event_hour_level_event_datetimeEpoch"])
```

```python
df_weather_v4.head()
```

```python
df_weather_v4.to_csv('data/df_weather+alarms_merged.csv')
```
